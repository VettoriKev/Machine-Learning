#Ejercicio 2

Creamos los dos dataframes con distribución normal.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
m1 = -20 ; m2 = 15
std1 = 25 ; std2 = 18 
N1 = 50 ; N2 = 50
 
A_val = np.random.normal(m1,std1,N1)
B_val = np.random.normal(m2,std2,N2)
 
A_ds = pd.DataFrame(columns=['Etiqueta', 'Valor'], index=range(N1))
A_ds['Valor'] = A_val
A_ds['Etiqueta'] = 'A'
 
B_ds = pd.DataFrame(columns=['Etiqueta', 'Valor'], index=range(N2))
B_ds['Etiqueta'] = 'B'
B_ds['Valor'] = B_val
```

### Vemos el primer dataset

```python
A_ds.sample(10)
```

```python
A_ds.plot.hist(bins=20)
A_ds.plot.density()
```

### Vemos el segundo dataset

```python
B_ds.sample(10)
```

```python
B_ds.plot.hist()
B_ds.plot.density()
```

```python
# @title Superponemos los histogramas de ambas listas

plt.hist(A_ds['Valor'], label = "A")
plt.hist(B_ds['Valor'], label = 'B')
plt.legend(labels = ['A','B'])
plt.show()
```

```python
# @title Concatenamos ambos histogramas

Comb_ds = pd.concat([A_ds,B_ds])
Comb_ds.sample(10)
```

```python
Comb_ds.plot.hist(bins=15)
Comb_ds.plot.density()
```

```python
# @title Ahora lo oredamos con respecto al valor de cada uno

Comb_Ord_ds = Comb_ds
Comb_Ord_ds.sort_values('Valor', inplace=True, ignore_index=True)
Comb_Ord_ds
```

## Procedemos a graficar nuestra curva ROC y la ideantidad.

```python
x_list, y_list = ([], [])
x_list.append(0)
y_list.append(0)

x, y = (0, 0)

for i in range(N1+N2):
    if Comb_Ord_ds.iloc[-(1+i), 0] == 'B':
        y += 1/N1
    else:
        x += 1/N2
    x_list.append(x)
    y_list.append(y)


plt.figure(1, (6,6))
plt.plot(x_list, y_list, [0,1], [0,1])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.grid()
plt.legend(['Test', 'Identidad'])
```

```python
# @title Creamos un dataframe para colocar los valores predictivos
Pred_ds = pd.DataFrame(columns=['Valor', 'Etiqueta_predecida', 'Etiqueta_real'], index=range(N1+N2))
Pred_ds['Valor'] = Comb_ds['Valor'] 
Pred_ds['Etiqueta_real'] = Comb_ds['Etiqueta']
Pred_ds.sample(5)
```

```python
# @title Calculamos el punto de maxima distancia a la recta de ideantidad para usar como umbral

dist = np.sqrt(np.abs(np.array(x_list)-((np.array(x_list)+np.array(y_list))/2))+np.abs(np.array(y_list)-((np.array(x_list)+np.array(y_list))/2))) 

[pos_umbral] =np.where(dist == dist.max())
pos_umbral = pos_umbral[-1]  #Elegi uno de los posibles valores, pero en el caso de que haya mas de un valor que tenga una distancia maxima, puede optarse cualquier criterio de eleccion entre ellos


Umbral = Comb_Ord_ds.iloc[-(1+pos_umbral),1]


Pred_ds['Etiqueta_predecida'] = ['A' if Pred_ds.iloc[i,0] < Umbral else 'B' for i in range(N1+N2)]
Pred_ds.sample(10)
```

```python
# @title Verificamos el tipo de error
Pred_ds['Tipo_de_error'] = ['FP' if Pred_ds.iloc[i,1]=='A' and Pred_ds.iloc[i,2]=='B' else 
                            'FN' if Pred_ds.iloc[i,1]=='B' and Pred_ds.iloc[i,2]=='A' else 
                            'VP' if Pred_ds.iloc[i,1]=='B' and Pred_ds.iloc[i,2]=='B' else
                            'VN' for i in range(N1+N2)]
Pred_ds
```

```python
# @title Calculamos las tazas de valores resultantes.

fp, fn, vp, vn = [np.sum(Pred_ds['Tipo_de_error'] == TdE) for TdE in ['FP', 'FN', 'VP', 'VN']]

TPR = vp/(vp+fn)
FPR = fp/(fp+vn)
PPV = vp/(vp+fp)
f_score = 2*TPR*PPV/(TPR+PPV)

print('TPR:', np.round(TPR, 3))
print('FPR:', np.round(TPR, 3))
print('PPV:', np.round(PPV, 3))
print('F-score:', np.round(f_score, 3))
```

### Vemos que con un metodo bastante sencillo, podemos elegir un identificador de umbral. Para tener una clasificación binaria. 
#### Claro, todo esto depende de que costo tiene nuestros diferentes tipos de errores, ya que en esta distinción equivocarnos de datasheet no daña nada siendo un experimento adimensional.
